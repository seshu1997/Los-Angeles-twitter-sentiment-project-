## Weekly Crime & Arrest Rate in LA, 2019

```python
import requests
```

```python
import numpy as np
```

```python
dates = list(np.arange(1,32,1))
```

```python
months = list(np.arange(1,13,1))
```

```python
def rate(mm,dd,name):
    rate = {}
    if name == 'crime':
        string = "https://data.lacity.org/resource/63jg-8b9z.json?date_occ=2019-" 
    else:
        string = 'https://data.lacity.org/resource/yru6-6re4.json?arst_date=2019-'
        
    for month in mm:
        
        new_str = string + str(month) + '-'
        
        if month < 10:
            mon = '2019-'+ '0' + str(month) + '-'
        else:
            mon = '2019-'+ str(month) + '-'
    
        for date in dd:
            url = new_str + str(date)
            if date < 10:
                title = mon +'0'+ str(date) 
            else:    
                title = mon + str(date)
            response = requests.get(url)
            rate[title] = len(response.json())
    
    return rate 
```

```python
crime_by_date = rate(months,dates,'crime')
```

```python
crime = pd.DataFrame(crime_by_date.items(),columns = ['Date','Incidence'])
```

```python
crime_2018 = len(requests.get("https://data.lacity.org/resource/63jg-8b9z.json?date_occ=2018-12-31").json())
```

```python
crime_2018
```

```python
crime.loc[-1] = ['2018-12-31',607] 
```

```python
crime.index = crime.index + 1
crime = crime.sort_index()
```

```python
arrest_by_date = rate(months,dates,'arrest')
```

```python
arrest = pd.DataFrame(arrest_by_date.items(),columns = ['Date','Incidence'])
```

```python
arrest_2018 = len(requests.get("https://data.lacity.org/resource/yru6-6re4.json?arst_date=2018-12-31").json())
arrest_2018
```

```python
arrest.loc[-1] = ['2018-12-31',238] 
arrest.index = arrest.index + 1
arrest = arrest.sort_index()
```

```python
arrest
```

```python
weeks = pd.read_csv('archive/week.csv')
weeks
```

```python
def weekly_rate(df,lst):
    sum_ = 0 
    dic = {}
    for i in range(len(weeks)): 
        new_df = df[(df['Date'] >= lst.iloc[i][1]) & (df['Date'] <= lst.iloc[i][2])]
        sum_ = sum(new_df['Incidence'])
        dic[i+1] = sum_
    return dic
```

```python
weekly_crime = weekly_rate(crime,weeks)
weekly_crime = pd.DataFrame(weekly_crime.items(),columns = ['Week #','Crime'])
weekly_crime
```

```python
weekly_crime.to_csv('weekly_crime.csv')
```

```python
weekly_arrest = weekly_rate(arrest,weeks)
weekly_arrest = pd.DataFrame(weekly_arrest.items(),columns = ['Week #','Arrest'])
weekly_arrest
```

```python
weekly_arrest.to_csv('weekly_arrest.csv')
```
